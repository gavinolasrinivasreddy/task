```python
threshold = 3
outlayaerdata=[]
def outlayer(data):
  std = np.std[data]
  mean = np.mean[data]
  for i in data:
    z = (i-mean)/std
    if np.abs(z)>3:
      outlayaerdata.append(i)
  return outlayaerdata
```

```python
outlayer(df['NumberOfTrips'])
```
