Train and Test split:

````python
from sklearn.model_selection import train_test_split
````

````python
X_train
````

````python
y_train
````

````python
X_test = scaler.transform(X_test)
````

````python
from sklearn.linear_model import LinearRegression
````

````python
reg = LinearRegression().fit(X_train, y_train)
````

````python
reg
````

coefficients


````python
print(reg.coef_)
````

Prediction
